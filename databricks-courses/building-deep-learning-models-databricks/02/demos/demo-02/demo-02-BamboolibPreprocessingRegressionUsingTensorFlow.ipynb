```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, MinMaxScaler

import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import EarlyStopping

import mlflow

from sklearn.metrics import r2_score
```

```python
X = car_details_df.drop(['selling_price'], axis = 1)

y = car_details_df['selling_price']

X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 123)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
y_train.to_csv('/dbfs/FileStore/datasets/processed_y_train.csv', index = False)
y_test.to_csv('/dbfs/FileStore/datasets/processed_y_test.csv', index = False)
```

Categorical features are listed and One hot encoder is defined

```python
categorical_features = ['fuel', 'seller_type', 'transmission', 'owner']

categorical_transformer = OneHotEncoder(drop = 'first')
```

ColumnTransformer : This estimator allows different columns or column subsets of the input to be transformed separately and the features generated by each transformer will be concatenated to form a single feature space. Here We are using Column Transformer to one hot encode categorical features defined above and Standard scale remaining numerical features

```python
preprocessor = ColumnTransformer(
    transformers = [('cat', categorical_transformer, categorical_features)], 
    remainder = StandardScaler()
)

print(preprocessor)
```

```python
X_train = pd.DataFrame(preprocessor.fit_transform(X_train),
                       columns =  preprocessor.get_feature_names_out(),
                       index = X_train.index)

X_test = pd.DataFrame(preprocessor.transform(X_test),
                      columns = preprocessor.get_feature_names_out(), 
                      index = X_test.index)
X_train.sample(5)
```

```python
X_train.to_csv('/dbfs/FileStore/datasets/processed_x_train.csv', index = False)
X_test.to_csv('/dbfs/FileStore/datasets/processed_x_test.csv', index = False)
```

Scaling the target using Min-max Scaler(as it ranges between 0 to 1 for positive input(price cant be negative)) as it has very high variance,which may result loss overflow

```python
min_max_scaler = MinMaxScaler()

y_train = min_max_scaler.fit_transform(y_train.values.reshape(-1, 1))

y_test = min_max_scaler.transform(y_test.values.reshape(-1, 1))

y_train
```

Let's define different models with different activation functions, different optimizers and learning rates

```python
def linear_activation_model():
    
    model = Sequential()
    model.add(Dense(16, input_dim = 16, activation = 'linear'))
    model.add(Dense(8, activation = 'linear'))
    model.add(Dense(1))
    
    model.compile(loss = 'mse',
              optimizer = tf.keras.optimizers.Adam(learning_rate = 0.0001),
              metrics = ['mse'])
    
    return model
```

```python
def relu_activation_model():
    
    model = Sequential()
    model.add(Dense(32, activation = 'relu'))
    model.add(Dense(16, input_dim = 16, activation = 'relu'))
    model.add(Dense(8, activation = 'relu'))
    model.add(Dense(1))
    
    model.compile(loss = 'mse',
                  optimizer = tf.keras.optimizers.SGD(learning_rate = 0.1),
                  metrics = ['mse'])
    
    return model
```

```python
def sigmoid_activation_model():
    
    model = Sequential()
    model.add(Dense(32, input_dim = 16, activation = 'sigmoid'))
    model.add(Dense(16, input_dim = 16, activation = 'sigmoid'))
    model.add(Dense(8, activation = 'sigmoid'))
    model.add(Dense(1))
    
    model.compile(loss = 'mse',
              optimizer = tf.keras.optimizers.SGD(learning_rate = 0.1),
              metrics = ['mse'])
    
    return model
```

This time let's also log the r-square
Use mlflow.log_metric

```python
mlflow.tensorflow.autolog()

with mlflow.start_run(run_name = 'linear_activation_network') as first_model_run:
    
    model = linear_activation_model()
    
    early_stopping = EarlyStopping(monitor = 'loss', mode = 'min', patience = 3)

    history = model.fit(X_train, y_train, validation_split = .2, epochs = 250, callbacks = [early_stopping])
    
    y_pred = model.predict(X_test)
    
    rsquare_score = r2_score(y_test, y_pred)
    mlflow.log_metric('rsquare_score', rsquare_score)
    
    fig, ax = plt.subplots(figsize = (12, 8))
 
    plt.scatter(y_test, y_pred)
 
    plt.xlabel('Actual')
    plt.ylabel('Predicted')

    mlflow.log_figure(fig, 'Predicted_vs_Actual.png')

    first_model_run = mlflow.active_run()
    print('Active run_id: {}'.format(first_model_run.info.run_id))
```
