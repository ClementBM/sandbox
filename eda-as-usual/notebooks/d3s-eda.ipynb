# Question 1

Classer les produits par chiffre d’affaires pour l’entreprise. Combien de produits représente 
80% du chiffre d’affaires ?

```python
price_df["CA_by_product_destination_country"] = price_df["Unit Price"] * price_df["Sales qty 2022"]
ca_df = price_df.groupby("Product_ID")["CA_by_product_destination_country"].sum().sort_values(ascending=False).to_frame()
```

```python
total_ca = ca_df["CA_by_product_destination_country"].sum()
ca_df["CA_pct"] = ca_df["CA_by_product_destination_country"] / total_ca
ca_df["CA_pct_cumsum"] = ca_df["CA_pct"].cumsum()
ca_df.query("CA_pct_cumsum <= 0.8")
```

# Question 3

Calculer la marge totale faite par chacun des pays vendeurs.

```python
price_cost_df = pd.merge(
    cost_df[['Product_ID', 'Production_Site', 'Unit Cost']],
    price_df[['Product_ID', 'Selling_Country', 'Destination_Country', 'Sales qty 2022', 'Unit Price']],
    how="right",
    on="Product_ID")
price_cost_df["ca"] = price_cost_df["Unit Price"] * price_cost_df["Sales qty 2022"]
price_cost_df["loss"] = price_cost_df["Unit Cost"] * price_cost_df["Sales qty 2022"]
price_cost_df["benefit"] = price_cost_df["ca"] - price_cost_df["loss"]
```

```python
price_cost_df.groupby("Selling_Country")["benefit"].sum().sort_values(ascending=False)
```

# Question 4

* Pour chaque pays vendeur, quel est le produit représentant les plus grosses ventes ?
* Quelle est la part des ventes réalisée par ce produit pour ce pays ?

```python
benefit_by_item_by_country = price_cost_df.groupby(["Selling_Country", "Product_ID"])["ca"].sum().to_frame()
benefit_by_item_by_country
```

```python
benefit_by_item_by_country.loc["Afghanistan"].sort_values("ca")
```

```python
df4 = benefit_by_item_by_country.reset_index().sort_values(
    ["Selling_Country", "ca"],
    ascending=[True, False]).groupby("Selling_Country").first()
df4
```

```python
price_cost_by_country = price_cost_df.groupby(
    ["Selling_Country"])["ca"].sum().to_frame().rename(columns={"ca": "total_benefit"})
```

```python
best_item_by_country = pd.merge(price_cost_by_country, df4, how="inner", on="Selling_Country")
best_item_by_country["prop"] = best_item_by_country["ca"] / best_item_by_country["total_benefit"]
best_item_by_country
```
