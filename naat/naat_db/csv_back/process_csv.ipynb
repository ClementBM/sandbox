# Ground Type

```python
ground_type_df = pd.read_csv("ground_type.csv")
ground_type_df
```

# Ground

```python
ground_df = pd.read_csv("ground.csv", header=None)
ground_df.rename(
    columns={
        0: "Id",
        1: "Name",
        2: "Ground_Type_Id",
    }).merge(ground_type_df, on="Ground_Type_Id").to_csv("grounds.csv")
```

# Agent Type

# Agent

```python
agent_df = pd.read_csv("agent.csv", header=None)
agent_df.rename(columns={0: "Id", 1: "Name", 2: "AgentUrl", 3: "Description", 4: "AgentTypeId"}).head()
```

# Appeal Type

```python
appeal_type_df = pd.read_csv("appeal_type.csv")
appeal_type_df
```

# Case Status

```python
casestatus_df = pd.read_csv("case_status.csv")
casestatus_df
```

# Case

```python
case_df = pd.read_csv("case.csv", header=None)
case_df.rename(
    columns={0: "Id", 1: "Title", 2: "FactsDigest", 3: "GoogleDriveFolder", 8: "FactsDate", 4: "CaseStatusId", 5: "AppealTypeId", 9: "Status"}
).merge(casestatus_df, on="CaseStatusId", how="left").merge(appeal_type_df, on="AppealTypeId",how="left").head()
```

# Decision Status

```python
decision_status_df = pd.read_csv("decision_status.csv", header=None)
decision_status_df = decision_status_df.rename(
    columns={0: "DecisionStatusId", 1: "StatusName"}
)[["DecisionStatusId", "StatusName"]]
decision_status_df
```

# Decision

```python
decision_df = pd.read_csv("decision.csv")
decision_df.rename(columns={"7": "DecisionStatusId"}).merge(decision_status_df, on="DecisionStatusId", how="left").to_csv("decisions.csv")
```

# Resource Type

```python
resourcetype_df = pd.read_csv("resource_type.csv", header=None)
resourcetype_df = resourcetype_df.rename(columns={0: "Resource_Type_Id", 1: "TypeName"})
resourcetype_df
```

# Resource

```python
resource_df = pd.read_csv("resource.csv", header=None)
resource_df.rename(
    columns={0: "Id", 1: "Name", 2: "Url", 3: "Resource_Type_Id"}
).merge(resourcetype_df, on="Resource_Type_Id")[["Name", "Url", "TypeName"]].head()
```
