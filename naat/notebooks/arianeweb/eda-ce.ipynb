```python
import os, io, tarfile
from zipfile import ZipFile
from artifacts import ARIANEWEB_PATH
import pandas as pd
```

```python
decision_folder = ARIANEWEB_PATH / "CE"
zip_files = []

for file_path in decision_folder.iterdir():
    if file_path.suffix == ".zip":
        zip_files.append(file_path)
```

```python
xml_decisions = []

# opening the zip file in READ mode
with ZipFile(zip_files[1], 'r') as zip:
    for file in zip.filelist:
        with zip.open(file) as file_reader:
            filedata = file_reader.read()
            xml_decisions.append(pd.read_xml(filedata))
```
